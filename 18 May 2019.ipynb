```python
import numpy as np

t = np.arange(0.0,2.0,0.1)
v = 1+np.sin(2 * np.pi * t)

plt.xlabel('time')
plt.ylabel('distance')
plt.grid(True)
plt.plot(t,v)
plt.savefig('time&voltage.png')
plt.show()

```

```python
import pandas as pd 
df = pd.read_csv('PythonScripts/Datafiles/income.csv')
y=df.values[4]
x=df.columns
x
y
```

```python
years=[]
for i in range(2,len(x)):
    years.append(x[i])
years
income=[]
for i in range(2,len(y)):
    income.append(y[i])
income
plt.xlabel('Year')
plt.ylabel('Income')
plt.grid(True)
plt.plot(years,income)
```

```python
width=1/3
plt.title('Income of California')
plt.bar(years,income,width,color='Gray')

```
